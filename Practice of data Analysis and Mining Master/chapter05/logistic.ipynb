## 降低贷款拖欠率

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression as LR
from sklearn.linear_model import RandomizedLogisticRegression as RLR
```

```python
# 使用随机逻辑回归进行特征筛选
data = pd.read_excel('data/bankloan.xls')
x = data.iloc[:, :8].as_matrix()
y = data.iloc[:, -1].as_matrix()
data.head()
```

```python
rlr = RLR()     # 建立随机逻辑回归模型筛选变量
rlr.fit(x, y)
```

```python
rlr.get_support()
```

```python
print(u'有效特征为：%s' % ','.join(data.columns[rlr.get_support()]))
```

```python
x = data[data.columns[rlr.get_support()]].as_matrix()  #筛选选择后的特征值
```

```python
lr = LR()
lr.fit(x, y)
print(u'模型的平均正确率为：%s' % lr.score(x, y))
```

>使用递归特征消除，其主要思想是反复构建模型（SVM或回归模型），然后选出最好的（或者最差的）特征（可以根据系数选择），把选出来的特征放到一边，在剩余的特征上重复这个过程,直到遍历所有特征。筛选掉的变量，不一定与结果无关，可能是非线性关系，所以还要根据实际背景对筛选结果进行分析。
