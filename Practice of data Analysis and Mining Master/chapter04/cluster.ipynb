```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('data/discretization_data.xls')
data.head()
```

```python
newdata = data[u'肝气郁结证型系数'].copy()
newdata.head(20)
```

```python
# newdata.plot(kind='scatter')
plt.figure(figsize=(10, 6))
plt.scatter(newdata.index, newdata)
```

```python
k = 4
```

```python
# 等宽离散化
d1 = pd.cut(newdata, k, labels=range(k))
```

```python
# 等频离散化
w = [1.0 * i/k for i in range(k+1)]
w = newdata.describe(percentiles = w)[4:4+k+1]
w[0] = w[0] * (1-1e-10)
d2 = pd.cut(newdata, w, labels=range(k))
```

```python
# kmeans聚类方法
from sklearn.cluster import KMeans
kmodel = KMeans(n_clusters = k, n_jobs = 4)   #n_jobs是并行数，一般等于cpu数较好
kmodel.fit(newdata.reshape( len(newdata),1))
```

```python
c = pd.DataFrame(kmodel.cluster_centers_).sort_values(0)
c
```

```python
# w  = pd.rolling_mean(c,2).iloc[1:]
w = c.rolling(window=2, center=False).mean().iloc[1:]
w  = [0] + list(w[0]) + [newdata.max()]
d3 = pd.cut(newdata, w, labels=range(k))
```

```python
import matplotlib.pyplot as plt
def cluster_plot(d, k):
    plt.figure(figsize = (8,3))
    for j in range(k):
        plt.plot(newdata[d==j], [j for j in d[d==j]], 'o')
        
    plt.ylim(-0.5, k - 0.5)
    return plt

cluster_plot(d1, k).show()
cluster_plot(d2, k).show()
cluster_plot(d3, k).show()
```
